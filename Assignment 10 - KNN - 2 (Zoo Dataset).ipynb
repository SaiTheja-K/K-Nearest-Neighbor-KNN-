### Model Building

```python
X = zoo_data.drop(['animal name','type'],axis=1)
y = zoo_data[['type']]
```

```python
X.shape,y.shape
```

```python
X.head()
```

```python
y.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,stratify=y,random_state=12121)
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

### finding the optimal number of neighbors.

```python
k_values = np.arange(1,25)
train_accuracy = []
test_accuracy = []
```

```python
for i, k in enumerate(k_values):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    train_accuracy.append(knn.score(X_train, y_train))
    test_accuracy.append(knn.score(X_test, y_test))
```

```python
plt.figure(figsize=[13,8])
plt.plot(k_values, test_accuracy, label = 'Testing Accuracy')
plt.plot(k_values, train_accuracy, label = 'Training Accuracy')
plt.legend()
plt.title('-value VS Accuracy')
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.xticks(k_values)
plt.show()
```

### As shown in the graph, with K=5,we can achieve train accuracy of 94% and achieve a test accuracy of 86%.

### Model Training

```python
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train,y_train)
```

### Model Testing & Model Evaluation

### Train Data

```python
y_pred_train = knn_model.predict(X_train)
pd.Series(y_pred_train).value_counts()
```

```python
acc_train= accuracy_score(y_train, y_pred_train)
print('Accuracy :', round(acc_train,2))
print('Classification Report : \n', classification_report(y_train, y_pred_train))
```

### Test Data

```python
y_pred_test = knn_model.predict(X_test)
```

```python
acc_test= accuracy_score(y_test, y_pred_test)
print('Accuracy :', round(acc_test,2))
print('Classification Report : \n', classification_report(y_test, y_pred_test))
```

### Conclusion with K=5, we got the accuracy of 86%

### So, the animals are categorized with 86% accuracy
